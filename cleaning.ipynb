```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('startup_funding.csv')
```

```python
df.head(3)
```

```python
sorted(df['Startup Name'].unique().tolist())
```

```python
sorted(df['Investors Name'].unique().tolist())
```

```python
df['Investors Name'].isnull().sum()
```

```python
df['Investors Name']=df['Investors Name'].fillna('undisclosed')
```

```python
df['Investors Name'].isnull().sum()
```

```python
sorted(df['Investors Name'].unique().tolist())
```

```python
df.info()
```

```python
df.drop(columns=['Remarks'],inplace=True)
```

```python
df.set_index('Sr No',inplace=True)
```

```python
df.sample(1)
```

```python
df.rename(columns={
    'Date dd/mm/yyyy':'date',
    'Startup Name':'startup',
    'Industry Vertical':'vertical',
    'SubVertical':'subverticle',
    'City  Location':'city',
    'Investors Name':'investors',
    'InvestmentnType':'round',
    'Amount in USD':'amount'
},inplace=True)
```

```python
df['amount']=df['amount'].fillna('0')
```

```python
df['amount'].isnull().sum()
```

```python
df['amount']=df['amount'].str.replace(',','')
df['amount']=df['amount'].str.replace('undisclosed','')
df['amount']=df['amount'].str.replace('unknown','')
df['amount']=df['amount'].str.replace('Undisclosed','')
```

```python
df=df[df['amount'].str.isdigit()]
```

```python
df['amount']=df['amount'].astype('float')
```

```python
df.info()
```

```python
def to_inr(doller):
    inr=doller * 90.97
    return inr/10000000
```

```python
df['amount']=df['amount'].apply(to_inr)
```

```python
df.head(3)
```

```python
df['date']=pd.to_datetime(df['date'],errors='coerce')
```

```python
df.shape
```

```python
df['date'].value_counts()
```

```python
df=df.dropna(subset=['date','startup','vertical','city','investors','round','amount'])
```

```python
df.shape
```

```python
df.info()
```

```python
df.to_csv('startup_funding_clean.csv',index=False)
```
